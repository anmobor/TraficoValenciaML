```python
todaslascalles = trafico_union_semestres['descripcion'].unique() #COGE TODAS LAS CALLES

for calle in todaslascalles:
  calle_y_fecha = trafico_union_semestres.loc[ #COGE BIEN SOLO LA CALLE QUE LE PONGO
            trafico_union_semestres['descripcion'] == calle, ['descripcion','fecha']]
  fechadf = pd.DataFrame({'fecha': calle_y_fecha['fecha'].unique()})
  calledf = pd.DataFrame({ 'descripcion': calle_y_fecha['descripcion'].unique() })
  fechacalle_dataframe= pd.concat( [ fechadf,  calledf], axis=1 )
  fechacalle_dataframe['descripcion'] = calle
#  ts = pd.DataFrame({'ds': calle_y_fecha['fecha'], 'y': calle_y_fecha.descripcion})
#  fechacalle_dataframe.rename( columns={0 :'fecha'}, inplace=True )
  timeseries = pd.Series(data=fechacalle_dataframe['descripcion'][0], index=fechacalle_dataframe['fecha'])
  timeseries.asfreq('1D')
  print('index0 inicio',timeseries.index[0])
  print('index size final', timeseries.index[timeseries.size-1])
  faltan = pd.date_range(start = timeseries.index[0], end = timeseries.index[timeseries.size-1] ).difference(timeseries.index)
  print('Dias que faltan: ',faltan)
  print('Numero de Dias que faltan: ',faltan.size) # Son saltos en las fechas de la serie temporal sin datos, huecos de datos
  print('Numero de Dias que hay: ',timeseries.size)
  print(calle)
  if faltan.size > 31 or timeseries.size < 1000:# si faltan más de 31 días en de los datos para la calle
      trafico_union_semestres.drop(trafico_union_semestres.loc[trafico_union_semestres['descripcion']==calle].index, inplace=True)

print(trafico_union_semestres.count())
trafico_union_semestres.to_pickle('tramosFechasCompletas.plk')
```
